# Training our model
 ---

- train.py is python file for training
- device is 0 
- batch size is 16
- epoch 100 
- img size will 640 X 640 (official yolo v7 suggest)
- custom_data.ymal will provide the path of training and test images and also number of classes
- And set its hyperparameters, from github.
- Initialize with Random weights 

```python
!python train.py --device 0 --batch-size 16 --epoch 100 --img 640 640 --data data/custom_data.yaml --hyp data/hyp.scratch.custom.yaml --cfg cfg/training/yolov7-custom.yaml --weights yolov7.pt --name yolov7-custom
```

Lets train further iteration with adding 200 epochs, on our best weights at.

runs/train/yolov7-custom2/weights/best.pt


```python
!python train.py --device 0 --batch-size 16 --epoch 200 --img 640 640 --data data/custom_data.yaml --hyp data/hyp.scratch.custom.yaml --cfg cfg/training/yolov7-custom.yaml --weights runs/train/yolov7-custom2/weights/best.pt --name yolov7-custom
```
